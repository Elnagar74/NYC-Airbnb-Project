### Airbnb Listings EDA Project - New York 2024

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading Dataset**

```python
df = pd.read_csv('Airbnb Dataset - 2024.csv', encoding_errors='ignore')
```

**Data Cleaning**

```python
df.isnull().sum()
```

```python
# Removing rows with missing values
df.dropna(inplace=True)
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
# dropping all duplicates rows
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

**Data Analysis**

```python
df.columns
```

**Univariate Analysis**

```python
# Price distribution
sns.histplot(data=df, x='price')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Removing outliers from the price column
df = df[df['price'] < 1500]
sns.boxplot(data=df, x='price')
plt.title('Boxplot of Price (Outliers Removed)')
plt.xlabel('Price')
plt.show()
```

```python
# Price distribution after removing outliers
sns.histplot(data=df, x='price', bins=50, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# availability_365 distribution
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribution')
plt.xlabel('Availability (days)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top 10 most frequent ratings
top_ratings = df['rating'].value_counts().nlargest(10)
sns.barplot(x=top_ratings.index, y=top_ratings.values)
plt.title('Top 10 Most Frequent Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

**Bivariate Analysis**

```python
# Relationship between price and beds
sns.scatterplot(data=df, x='beds', y='price')
plt.title('Price vs Beds')
plt.xlabel('Number of Beds')
plt.ylabel('Price')
plt.show()
```

```python
# Comparison of average prices by neighbourhood group and room type.
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.title('Average Price by Neighbourhood Group and Room Type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.legend(title='Room Type')
plt.show()
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365', 'beds'], hue='room_type')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
# Geographical Distribution of AirBnb Listings By Room Type
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type', palette='colorblind')
plt.title("Geographical Distribution of AirBnb Listing By Room Type")
plt.show()
```

```python
# Geographical Distribution of AirBnb Listings By Beds
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='beds', palette='colorblind')
plt.title("Geographical Distribution of AirBnb Listing By Room Type")
plt.show()
```

```python
df[['id', 'host_id']] = df[['id', 'host_id']].astype(object)
df[['id', 'host_id']].dtypes
```

```python
# Correlation heatmap for all numeric features
Numric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
corr = df[Numric_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
